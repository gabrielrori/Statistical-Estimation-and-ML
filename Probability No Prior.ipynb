# Probability Distributions with and without Prior Information 

### Probabilidad sin información a priori de $\mu$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
NS = 1000 # Numero de samples
u_vector = np.linspace(6,14,num=NS)
```

```python
argumento_exp_m1 = (u_vector**2 - 20.6*u_vector +106.09)/2
argumento_exp_m1_m2 = u_vector**2 - 20.4*u_vector +104.04
p_u_m1 = 0.6033*np.exp(-argumento_exp_m1)
#p_u_m1_m2 = 0.59271*np.exp(-argumento_exp_m1_m2)
p_u_m1_m2 = 0.68286*np.exp(-argumento_exp_m1_m2)
```

```python
plt.plot(u_vector, p_u_m1)
plt.grid('on')
plt.xlabel('$\mu$')
plt.ylabel('$P(\mu|m)$')
```

```python
plt.plot(u_vector, p_u_m1_m2)
plt.grid('on')
plt.xlabel('$\mu$')
plt.ylabel('$P(\mu|m)$')
```

### Se obtiene una probabilidad que no es acotada debido a que no hay información de la variable a estimar

### Probabilidad con información a priori de $\mu$

```python
p_u_m1 = (u_vector > 9) * p_u_m1
p_u_m1 = (u_vector < 11) * p_u_m1

p_u_m1_m2 = (u_vector > 9) * p_u_m1_m2
p_u_m1_m2 = (u_vector < 11) * p_u_m1_m2
```

```python
plt.plot(u_vector, p_u_m1)
plt.grid('on')
plt.xlabel('$\mu$')
plt.ylabel('$P(\mu|m)$')
```

```python
plt.plot(u_vector, p_u_m1_m2)
plt.grid('on')
plt.xlabel('$\mu$')
plt.ylabel('$P(\mu|m)$')
```

### Distribución de probabilidad a posteriori con información a priori, para este caso es acotada y resulta en mejores estimadores ya que tienen información a priori.

### Además, en general, mientras más mediciones se tengan disminuirá la varianza y la media se acercará a 10.
